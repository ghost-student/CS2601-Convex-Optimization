```python
import numpy as np
import gd
import utils
```

```python
def run(gamma, stepsize, maxiter=100000):
    Q = np.diag([1, gamma])
    x0 = np.array([1.0, 1.0])

    f = lambda x: 0.5 * x@Q@x
    fp = lambda x: Q@x
    f_2d = lambda x1, x2: 0.5 * x1**2 + 0.5 * gamma * x2**2

    x_traces = gd.gd_const_ss(fp, x0, stepsize, maxiter=maxiter)

    print('gamma = {}, step size = {}'.format(gamma, stepsize))
    print('number of iterations: {}'.format(len(x_traces) - 1))

    utils.plot_traces_2d(f_2d, x_traces, 'figures/gd_traces_gamma{}_ss{}.pdf'.format(gamma, stepsize))
    utils.plot(f, x_traces, 'figures/gd_f_gamma{}_ss{}.pdf'.format(gamma, stepsize))
    print()
```

```python
# (b) Run the gradient descent algorithm with different step sizes
gamma = 10
stepsizes = [0.22, 0.1, 0.01, 0.001]

for stepsize in stepsizes:
    if (stepsize == 0.22):
        run(gamma, stepsize, 10)
    else:
        run(gamma, stepsize)
```

```python
# (c) Run the gradient descent algorithm with different gammas
stepsize = 1
gammas = [1, 0.1, 0.01, 0.001]

for gamma in gammas:
    run(gamma, stepsize)
```

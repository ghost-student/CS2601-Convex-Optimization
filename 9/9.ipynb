```python
import numpy as np
import matplotlib.pyplot as plt
import newton as nt
import utils
```

## Initialize the problem

```python
def f(x):
    #   START OF YOUR CODE
    return x - np.log(x) if x > 0 else float('inf')
    #   END OF YOUR CODE

def fp(x):
    #   START OF YOUR CODE
    return 1 - 1 / x
    #   END OF YOUR CODE

def fpp(x):
    #   START OF YOUR CODE
    return 1 / x ** 2
    #   END OF YOUR CODE
```

```python
# use the value you find in part (a)
f_opt = 1
def error(x):
    return f(x) - f_opt
```

## Try running the Newton step

Part (c)

```python
for x0 in [0.5, 1.5, 2.5]:
    x_traces = nt.newton(fp, fpp, x0, maxiter=5)
    for i, x in enumerate(x_traces):
        print("x_{} = {}".format(i, x))
    print()
```

## Damped Newton's method

Part (e)

```python
for x0 in [0.5, 1.5, 2.5]:
    x_traces, stepsize_traces = nt.newton_damped(f, fp, fpp, x0, alpha=0.1, beta=0.7)
    f_value = f(x_traces[-1])

    for i, x in enumerate(x_traces):
        print("x_{} = {}".format(i, x))
    print()

    print('Damped Newton\'s method')
    print('number of iterations: {}'.format(len(x_traces) - 1))
    print('solution: {}'.format(x_traces[-1]))
    print('value: {}'.format(f_value))
    print()

    utils.plot_stepsizes(stepsize_traces, 'figures/nt_damped_ss_{}.pdf'.format(x0))
    utils.plot(error, x_traces, 'figures/nt_error_damped_{}.pdf'.format(x0))
```
